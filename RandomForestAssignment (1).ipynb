In this assignment students will build the random forest model after
normalizing the variable to house pricing from boston data set.

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.model_selection import train_test_split,GridSearchCV 
from sklearn.preprocessing import StandardScaler 
from sklearn import datasets 
from sklearn.ensemble import RandomForestRegressor
boston = datasets.load_boston()
features = pd.DataFrame(boston.data,columns=boston.feature_names)
targets = boston.target
```

```python
features.head()
```

```python
features.describe()
```

```python
boston
```

```python
target=boston.target
features.shape
```

```python
sc=StandardScaler()
```

```python
X=sc.fit_transform(features)
Y=sc.fit_transform(target.reshape(-1,1))
Y
```

```python
x_train,x_test,y_train,y_test=train_test_split(X,Y,test_size=0.2,random_state=300)
```

```python
rfr=RandomForestRegressor(oob_score=True)
```

```python
rfr.fit(x_train,y_train.ravel())
```

```python
rfr.oob_score_
```

```python
rfr.get_params()
```

```python
#n_estimators : This is the number of trees you want to build before taking the maximum voting or averages of predictions. 
grid_param={
    'n_estimators':[50,100,150,200],
    'max_depth':range(2,10),
    'max_features':['auto','log2'],
    
}
```

```python
grid_search=GridSearchCV(estimator=rfr,param_grid=grid_param,cv=5,n_jobs=-1,verbose=3)
```

```python
grid_search.fit(x_train,y_train)
```

```python
grid_search.best_params_
```

```python
grid_search.best_score_
```

```python
from sklearn.model_selection import cross_val_score
```

```python
score_before_tuning=cross_val_score(rfr,x_train,y_train.ravel(),cv=5).mean()
score_before_tuning
```

```python
rfr1=RandomForestRegressor(max_depth=9, max_features= 'log2', n_estimators= 100,oob_score= True,random_state=6)
```

```python
rfr1.fit(x_train,y_train.ravel())
```

```python
rfr1.oob_score_
```

```python
score_after_tuning=cross_val_score(rfr1,x_train,y_train.ravel(),cv=5).mean()
score_after_tuning
```

```python
test_score_before_tuning=rfr.score(x_test,y_test)
test_score_before_tuning
```

```python
test_score_after_tuning=rfr1.score(x_test,y_test)
test_score_after_tuning
```
